# Import libarires

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Read CSV file

```python
df = pd.read_csv("ngaussian.csv", delimiter=' ', header=None)
df.columns = ['X1', 'X2']
```

# Scatter all data

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['X1'], df['X2'], label='Data')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.title('Scatter Plot: X1 vs. X2')
```

# Fit X2 on X1

```python
regression1 = LinearRegression(fit_intercept=False)
X1 = df['X1'].values.reshape(-1, 1)
X2 = df['X2'].values.reshape(-1, 1)
regression1.fit(X1, X2)
B12 = regression1.coef_[0][0]
```

```python
print(f"B12 is {B12}")
```

```python
# Scatter plot with regression line for X2 on X1
plt.figure(figsize=(10, 5))
plt.scatter(X1, X2, label='Data')
plt.plot(X1, regression1.predict(X1), color='red', label=f'X2 = {B12} * X1')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.title('Scatter Plot: X1 vs. X2 with Regression Line (X2 on X1)')
plt.show()
```

```python
# Create a scatter plot for X1 and X2 - B12 * X1
plt.figure(figsize=(10, 5))
plt.scatter(X1, X2 - B12 * X1, label=f'X2 - {B12} * X1')
plt.xlabel('X1')
plt.ylabel('X2 - B12 * X1')
plt.legend()
plt.title('Scatter Plot: X1 vs. X2 - B12 * X1')
plt.show()
```

# Fit X1 on X2

```python
regression2 = LinearRegression(fit_intercept=False)
regression2.fit(X2, X1)
B21 = regression2.coef_[0][0]
```

```python
print(f"B21 is {B21}")
```

```python
# Scatter plot with regression line for X1 on X2
plt.figure(figsize=(10, 5))
plt.scatter(X2, X1, label='Data')
plt.plot(X2, regression2.predict(X2), color='red', label=f'X1 = {B21} * X2')
plt.xlabel('X2')
plt.ylabel('X1')
plt.legend()
plt.title('Scatter Plot: X2 vs. X1 with Regression Line (X1 on X2)')
plt.show()
```

```python
# Create a scatter plot for X2 and X1 - B21 * X2
plt.figure(figsize=(10, 5))
plt.scatter(X2, X1 - B21 * X2, label=f'X1 - {B21} * X2')
plt.xlabel('X2')
plt.ylabel('X1 - B21 * X2')
plt.legend()
plt.title('Scatter Plot: X2 vs. X1 - B21 * X2')
plt.show()
```
